##**Top selling product in each category**

```python
  import pandas as pd


  df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')

  # Group by category and find the top-selling product in each category
  top_selling_products = df.groupby('category')['product_name', 'quantity'].agg(
      top_selling_product=('product_name', lambda x: x.mode().iat[0]),
      total_quantity=('quantity', 'sum')
  ).reset_index()

  # Display the top-selling products in each category
  print(top_selling_products)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='category', y='total_quantity', hue='top_selling_product', data=top_selling_products)
plt.title('Top Selling Product in Each Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Top Selling Product', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()

```

##**Category wise total sales revenue**

```python
import pandas as pd

#  Read the dataset
df = pd.read_csv('/content/drive/MyDrive/retail_dataset.csv')

# Group the data by "Product Category"
grouped = df.groupby('category')

#  Calculate total sales revenue for each category and total quantity sold for each product
category_sales = grouped['quantity', 'total_amount'].sum().reset_index().rename(columns={'quantity': 'Total Quantity Sold', 'total_amount': 'Total Sales Revenue'})
# category_sales
product_sales = df.groupby('product_name')['quantity'].sum().reset_index().rename(columns={'quantity': 'Total Quantity Sold'})

#  Identify the top-selling products in terms of quantity sold
top_selling_products = product_sales.sort_values(by='Total Quantity Sold', ascending=False).head(5)


print("Total Sales Revenue by Product Category:")
print(category_sales)

print("\nTop-Selling Products in Terms of Quantity Sold:")
print(top_selling_products)


sns.set(style="whitegrid")

# Create a color palette
colors = sns.color_palette("pastel")

# Create a bar plot with color customization
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x='category', y='Total Sales Revenue', data=category_sales, palette=colors)

# Customize the plot
ax.set_xlabel('Category', fontsize=14)
ax.set_ylabel('Total Sales Revenue', fontsize=14)
ax.set_title('Total Sales Revenue by Category', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Add data labels on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():,.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)

# Display the plot
plt.show()


```

###**Top selling product**

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Set a seaborn style
sns.set(style="whitegrid")

# Create a color palette
colors = sns.color_palette("viridis")

# Create a bar plot with color customization and decreased width
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x='product_name', y='Total Quantity Sold', data=top_selling_products, palette=colors, width=0.6)

# Customize the plot
ax.set_xlabel('Product Name', fontsize=14)
ax.set_ylabel('Total Quantity Sold', fontsize=14)
ax.set_title('Top Selling Products', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Add data labels on top of bars
for p in ax.patches:
    ax.annotate(f"{int(p.get_height()):,}", (p.get_x() + p.get_width() / 4., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```

# How do sales vary over different months or seasons?
# Are there any noticeable trends or seasonal patterns in sales?

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/drive/MyDrive/retail_dataset.csv', parse_dates=['date'])

# Step 2: Extract month and year from the 'date' column
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year


monthly_sales = df.groupby(['year', 'month'])['total_amount'].sum().reset_index()
monthly_sales
```

```python

import plotly.express as px

# Assuming you have the DataFrame named 'monthly_sales' with columns 'year', 'month', and 'total_amount'
fig = px.line(monthly_sales,
              x=monthly_sales['month'],
              y=monthly_sales['total_amount'],
              color=monthly_sales['year'],
              title='Monthly Sales Trend (2022-2023)',
              labels={'total_amount': 'Total Sales Amount', 'month': 'Month'},
              line_shape='spline',  # You can customize the line shape (options: 'linear', 'spline', 'hv', 'vh', 'hvh', 'vhv')
              markers=True,         # Show markers on data points
              template='plotly_dark',  # You can choose different templates, like 'plotly', 'ggplot2', 'seaborn', etc.
              )

fig.update_layout(xaxis=dict(tickmode='array', tickvals=list(range(1, 13)), ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']))

fig.show()

```


Sales Trend Summary:

1. **Sales ups and downs**:
   - Sales go up and down over the months.
   - Some months have more sales, and others have less.

2. **Best months:**
   - May and August are the best months for sales.

3. **Comparing years:**
   - When we compare the same months in 2022 and 2023:
     - March was better in 2022.
     - August was better in 2023.

4. **Consistency:**
   - Overall, sales are kind of steady.
   - Some months are always good or not so good.

5. **No big change:**
   - There's no big increase or decrease in sales overall.

6. **Important months:**
   -  May and August—they're important for sales.



# Location-Based Analysis:

```python
from pandas.core.indexes import category
import plotly.express as px

fig = px.bar(city_sales, x=city_sales['City'], y=city_sales['Total Sales'], color=city_sales['category'], title="City wise sales analysis")
fig.show()
```

From the provided data, we can make several observations:

2. **Grocery Category:**

3. **Beauty Category:**
